```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7,8,9,8,6,5,9,9]
temperaturee=[23.3,34,22,33,22,35,36,23,26,30,21,29,36,40]
```

```python
df=pd.DataFrame({"days":days,"temperaturee":temperaturee})
sns.lineplot(x="days",y="temperaturee",data=df,)
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
char_1=sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex")
char_1
```

```python
char_1=sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex",markers=("o","<"))
char_1
plt.show()
```

```python
char_1=sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex")
char_1
plt.title("lineplot",fontsize=20)
plt.xlabel("size",fontsize=15)
plt.ylabel("total_bill",fontsize=15)
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(12,6))
char_1=sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex")
char_1
plt.title("lineplot",fontsize=20)
plt.xlabel("size",fontsize=15)
plt.ylabel("total_bill",fontsize=15)
plt.show()
```
