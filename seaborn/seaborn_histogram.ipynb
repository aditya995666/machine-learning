```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.distplot(tips_df["size"],bins=50)
```

```python
sns.distplot(tips_df["size"],bins=50,kde=True)
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"],bins=50,kde=True,color="r")
plt.legend("total_bill")
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"],bins=bins,kde=True,color="r")
plt.title("tips")
plt.xticks(bins)
plt.legend("total_bill",frontsize=10)
```
