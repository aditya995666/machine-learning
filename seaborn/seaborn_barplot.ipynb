```python
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
import numpy as np
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,hue=tips_df.sex)
```

```python
order=['sun','mon','tue','wed']
sns.barplot(x='day',y='total_bill',hue='sex',
            data=tips_df)
```

```python
order=['sun','mon','tue','wed']
sns.barplot(x='day',y='total_bill',hue='sex',
            data=tips_df,order = order,estimator=np.max,,)
plt.show()
```

```python
sns.set()
kwargs={"alpha":0.9,"linewidth":5,"linestyle":":","edgecolor":"k"}
sns.barplot(x='day',y='total_bill',hue='sex',            data=tips_df,**kwargs,,)
```

```python
kwargs={"alpha":0.9,"linewidth":5,"linestyle":":","edgecolor":"k"}
sns.barplot(x='day',y='total_bill',hue='sex',
            data=tips_df,**kwargs)
```

```python
kwargs={"alpha":0.9,"linewidth":5,"linestyle":":","edgecolor":"k"}
sns.barplot(x='day',y='total_bill',
            data=tips_df,**kwargs)
plt.title("barplot of tips",fontsize=15)
plt.xlabel("days")
plt.ylebel("total_bill")
plt.show()
```
