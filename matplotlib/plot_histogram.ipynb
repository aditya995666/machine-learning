```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(20,45,(50))
py_studeent=np.random.randint(18,40,(60))
```

```python
print(ml_student_age,py_studeent)
```

```python
from matplotlib import style
plt.figure(figsize=(12,7))
style.use("ggplot")
plt.hist([ml_student_age,py_studeent],linestyle="--",color=["y","m"],rwidth=0.8,label=["ml_student_age","py_studeent"],edgecolor="r",linewidth=2)
plt.title("student age histogram",fontsize=20)

plt.xlabel("student categries",fontsize=20)
plt.ylabel("ml_student_age",fontsize=20)

plt.legend()
plt.show()
```
