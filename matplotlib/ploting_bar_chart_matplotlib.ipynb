```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
class1=['py','english','hindi','math','java']
class_student=[12,34,13,56,17]
class1_student=[34,23,45,67,19]
```

```python
plt.bar(class1,class_student,width=0.2,color='g',edgecolor="r",linewidth=3)
plt.xlabel("subject_name",fontsize=10)
plt.ylabel("present_student in class room",fontsize=10)
plt.title("student bar chart")
plt.show()
```

```python
plt.barh(class1,class_student,color='g',edgecolor="r",linewidth=3)
plt.xlabel("subject_name",fontsize=10)
plt.ylabel("present_student in class room",fontsize=10)
plt.title("student bar chart")
plt.show()
```

```python
student_index=np.arange(len(class1))
width=0.5
plt.bar(student_index,class_student,width,color='g',edgecolor="r",linewidth=3)

plt.bar(student_index+width,class1_student,width,color='y',edgecolor="k",linewidth=3)
plt.xlabel("subject_name",fontsize=10)
plt.ylabel("present_student in class room",fontsize=10)
plt.title("student bar chart")
plt.show()
```
